```python
import requests
from bs4 import BeautifulSoup 
```

```python
url="https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=117;type=trophy"
data=requests.get(url)
mydata=data.text
```

```python
data
```

```python
mydata
```

```python
import pandas as pd
```

```python
soup=BeautifulSoup(mydata,'html.parser')
```

```python
table=soup.find(class_='engineTable')
decompose_tr=soup.find_all("tr",class_='note')
for decompose in decompose_tr:
    decompose.decompose()
df=pd.read_html(str(table))
```

```python
wholedata=pd.concat(df)
```
